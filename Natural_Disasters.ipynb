# **1) HEAT MAP - Shows Null values in Dataset**

A **Heat Map** is a data visualization type where the individual values contained in a matrix through variations in coloring.

Fractal maps and tree maps both often use a similar system of color-coding to represent the values taken by a variable in a hierarchy.

```python
plt.figure(figsize=(15,7))

sns.heatmap((df.isnull()) | (df == 0), cmap = 'mako')

plt.title('null, missing and \'0\' values heatmap')
plt.show()
```

# **2) SWARM PLOT - Plotting Total Disaster's per Country in 2021**

A **Swarm Plot** is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes.

It plots one dot for each data item

```python
y=df.groupby('Year')        # groupby()- To group df based on Year
y
```

```python
for Year,Year_df in y:           # Iterate through he Year column and their corresponding df
  print(Year)                    # Prints Key Value - Year
  print(Year_df)                 # Prints df corresponding to the key as value
```

```python
gg=y.get_group(2021)       # To access a speciic df with key value 2021
gg
```

```python
yc=gg[['Year','Country']]              # Display countries wrt year 2021
yc
```

```python
plt.figure(figsize=(100,30))                                            # Width =100 and Height =30
plt.style.use('dark_background')

sns.swarmplot(x ='Country', data = yc,s=40).set_title('Swarm Plot')     # sns - Seaborn
```

# **4) STRIP PLOT- Plotting Top 10 Highest Total Affected Count wrt Countries**  

A **Strip Plot** is a single-axis scatter plot that is used to visualise the distribution of many individual one-dimensional values. 
 
 The values are plotted as dots along one unique axis, and the dots with the same value can overlap. 


```python
TA=df.sort_values(by='Total Affected',ascending=False).head(10)      # Sort Total Affected column in Descending order and print first 10 values

print(TA['Country'])          # Prints first 10 Countries with Highest Total Affected values
print(TA['Total Affected'])   # Prints first 10 Highest Total Affected values
```

```python
plt.figure(figsize=(20,6))

TA=df.sort_values(by='Total Affected',ascending=False).head(40)   # Sort Total Affected column in Descending order and print first 40 values

t=TA['Total Affected']
c=TA['Country']
plt.style.use('dark_background')

sns.stripplot(x=c,y=t,hue='Country',data=TA,s=9)      # x axis - Countries , y axis - Total Affected Count 
```

# **9) WORD CLOUD- Plotting Natural Disasters subtype**

A **Word Cloud** (also called tag cloud or weighted list) is a visual representation of text data. 

Words are usually single words, and the importance of each is shown with font size or color.

```python
import matplotlib.pyplot as plt 
from wordcloud import WordCloud, STOPWORDS
import pandas as pd

words = '' 
stopwords = set(STOPWORDS) 

#Iterating disaster subtype values
for val in df['Disaster Subtype'][0:3000]: 
     val = str(val) 
     tokens = val.split() 
#converting to lowercase
     for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 

     words += " ".join(tokens)+" "

#generate wordcloud
wc = WordCloud(width = 500, height = 500, background_color ='black', stopwords = stopwords, min_font_size = 10).generate(words) 
plt.figure(figsize = (15,7)) 

#display wordcloud
plt.imshow(wc) 
plt.axis("off") 
plt.show() 
```
